## 1. Meet `input()` (5–15 min)

`input()` pauses your program, waits for the user to type something and press Enter, and gives you back whatever they typed.

```python
favorite_color = input("What is your favorite color? ")
print(f"Cool, {favorite_color} is a great color!")
```

Notice the text inside `input("...")` — that's the **prompt**, the message shown to the user so they know what to type. Always include a clear prompt; a bare `input()` with no message leaves the user guessing.

⚠️ In this notebook environment, when a cell is waiting for input, you'll usually see a text box appear — type your answer there and press Enter (rather than typing into the code itself).

## 2. `input()` Always Gives You a String (15–30 min)

This is the single most important thing to remember about `input()`: **whatever the user types comes back as a string** — even if they type a number!

```python
age = input("How old are you? ")
print(type(age))   # <class 'str'> - even if you typed "15"!
```

If you try to do math directly with it, you'll hit the same `TypeError` from Lesson 2:

```python
age = input("How old are you? ")
next_year = age + 1
print(next_year)
```

To fix it, convert the string to a number using `int()` (for whole numbers) or `float()` (for decimals), just like you learned in Lesson 2:

```python
age = input("How old are you? ")
age = int(age)         # convert string to integer

next_year = age + 1
print(f"Next year you'll be {next_year}!")
```

💡 A common shortcut is to convert immediately, in the same line as the `input()` call:

```python
age = int(input("How old are you? "))
print(f"In 10 years you'll be {age + 10}!")
```

## 3. Building Interactive Mini-Programs (30–40 min)

Let's combine `input()` with everything from Lessons 1–4 to build something real: a simple **BMI-style calculator** (using a made-up formula for practice).

```python
name = input("What is your name? ")
weight = float(input("Enter your weight in kg: "))
height = float(input("Enter your height in meters: "))

bmi = weight / (height ** 2)
bmi_rounded = round(bmi, 1)

print(f"Hi {name}, your calculated value is {bmi_rounded}.")
```

Here's another one: a simple **restaurant bill splitter**.

```python
bill = float(input("What was the total bill? $"))
people = int(input("How many people are splitting it? "))

each_pays = bill / people
each_pays_rounded = round(each_pays, 2)

print(f"Each person owes ${each_pays_rounded}")
```

## 4. Common Pitfalls with `input()` (40–48 min)

### Pitfall 1 — Forgetting to convert
If you forget `int()`/`float()`, math operations will either crash (`+`) or do something unexpected (like `*` repeating a string):

```python
repeats = input("How many times? ")   # forgot int()!
print("Ha" * repeats)
```

That raises a `TypeError` — you can multiply a string by an **integer** (like `"Ha" * 3`) to repeat it, but not by another **string**, and `repeats` is still a string here since we never converted it. Try fixing it by converting `repeats` to an `int` first.

### Pitfall 2 — Trusting `input()` will always be a "clean" number
If the user types something that isn't a valid number (like `"twelve"` instead of `"12"`), `int()` will crash with a `ValueError`. We'll learn how to gracefully handle that in Lesson 21 (Errors & Debugging) — for now, just be aware it can happen.

```python
age = int(input("Enter your age as a number: "))
print(f"You are {age} years old.")
# Try typing "twelve" instead of "12" when this runs, and see what error you get!
```

### Pitfall 3 — Extra spaces
Sometimes users accidentally add spaces before/after their answer, which can cause comparisons to fail later (we'll see this matter a lot starting Lesson 7). A good habit is to `.strip()` string input right away:

```python
color = input("Favorite color? ").strip().lower()
print(f"You said: '{color}'")
```
