## 1. Arithmetic Operators (5–20 min)

Python supports all the math operators you'd expect, plus a couple of extras.

```python
print(5 + 3)    # addition
print(5 - 3)    # subtraction
print(5 * 3)    # multiplication
print(5 / 3)    # division -- always gives a float, even if it divides evenly
print(6 / 3)    # still a float: 2.0, not 2
```

There's also **exponentiation** (powers), using `**`:

```python
print(2 ** 3)    # 2 to the power of 3 = 8
print(5 ** 2)    # 5 squared = 25
print(10 ** 0.5) # square root of 10 (raising to the power of 0.5)
```

### Mixing int and float
When you do math between an `int` and a `float`, the result is always a `float` — Python "upgrades" it automatically so no precision is lost:

```python
print(type(5 + 3))       # int + int = int
print(type(5 + 3.0))     # int + float = float
print(type(5 / 2))       # division always gives float
```

## 2. Order of Operations (20–30 min)

Python follows the same math rules you learned in school — often remembered as **PEMDAS**: Parentheses, Exponents, Multiplication/Division, Addition/Subtraction (left to right).

```python
print(2 + 3 * 4)     # multiplication happens first -> 2 + 12 = 14
print((2 + 3) * 4)   # parentheses force addition first -> 5 * 4 = 20

print(10 - 2 ** 2)   # exponent first -> 10 - 4 = 6
print((10 - 2) ** 2) # parentheses first -> 8 ** 2 = 64
```

💡 **When in doubt, use parentheses.** Even when they're not strictly required, they make your intention crystal clear to anyone reading your code (including future you).

```python
# These give the same result, but the second is much easier to read at a glance
result_a = 100 * 0.08 + 5
result_b = (100 * 0.08) + 5
print(result_a, result_b)
```

## 3. The Modulus Operator `%` and Floor Division `//` (30–42 min)

`%` (modulus) gives you the **remainder** after division. It's one of the most useful operators in programming, especially for checking even/odd, or "does this divide evenly.

```python
print(7 % 2)    # 7 divided by 2 is 3 remainder 1 -> prints 1
print(10 % 2)   # 10 divided by 2 is 5 remainder 0 -> prints 0
print(9 % 3)    # divides evenly -> 0
print(9 % 4)    # 9 / 4 = 2 remainder 1 -> prints 1
```

### Even or odd?
A number is **even** if `number % 2 == 0` (no remainder), and **odd** otherwise. We'll use this constantly once we hit loops and conditionals.

```python
number = 17
print(number % 2)   # 1 means odd, 0 would mean even
```

`//` (floor division) gives you the division result **rounded down** to the nearest whole number — useful when you want "how many whole times does this fit":

```python
print(7 // 2)     # 3 (7/2 is 3.5, rounded down to 3)
print(10 // 3)    # 3
print(20 // 6)    # 3

# A classic combo: minutes to hours and minutes
total_minutes = 135
hours = total_minutes // 60
minutes = total_minutes % 60
print(f"{total_minutes} minutes = {hours} hours and {minutes} minutes")
```

## 4. Rounding Numbers (42–48 min)

`round()` rounds a number to the nearest whole number (or to a chosen number of decimal places):

```python
print(round(3.4))     # 3
print(round(3.6))     # 4
print(round(3.5))     # careful - Python rounds .5 to the nearest EVEN number, so this gives 4

print(round(3.14159, 2))   # round to 2 decimal places -> 3.14
print(round(19.995, 2))    # round to 2 decimal places -> 20.0 (approx, due to how floats are stored)
```

Remember from Lesson 2: `int()` **truncates** (just chops off the decimal), while `round()` actually rounds:

```python
value = 7.8
print(int(value))     # 7 - just chops it off
print(round(value))   # 8 - properly rounds
```
