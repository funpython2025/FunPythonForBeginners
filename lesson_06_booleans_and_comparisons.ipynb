## 1. Meet the Boolean Type (5–15 min)

Back in Lesson 2 you saw the `bool` type briefly. A **boolean** can only ever be one of two values: `True` or `False` (note the capital letters — this matters in Python!).

```python
is_raining = True
is_sunny = False

print(is_raining)
print(type(is_raining))
```

Booleans might seem simple, but they're the backbone of every decision a program makes. Any question with a yes/no answer becomes a boolean in code: "Is the user old enough?", "Did they win?", "Is the cart empty?"

Booleans also behave like numbers under the hood: `True` acts like `1`, and `False` acts like `0`.

```python
print(True + True)     # 2
print(True + False)    # 1
print(False * 10)      # 0
```

## 2. Comparison Operators (15–30 min)

Comparison operators compare two values and produce a boolean result — `True` or `False`.

| Operator | Meaning | Example | Result |
|----------|---------|---------|--------|
| `==` | equal to | `5 == 5` | `True` |
| `!=` | not equal to | `5 != 3` | `True` |
| `>`  | greater than | `7 > 10` | `False` |
| `<`  | less than | `7 < 10` | `True` |
| `>=` | greater than or equal to | `5 >= 5` | `True` |
| `<=` | less than or equal to | `4 <= 3` | `False` |

```python
print(5 == 5)
print(5 != 3)
print(7 > 10)
print(7 < 10)
print(5 >= 5)
print(4 <= 3)
```

You can compare variables just as easily as literal values:

```python
age = 15
voting_age = 18

print(age >= voting_age)
print(age < voting_age)
```

You can even store the *result* of a comparison in a variable, since it's just a value like any other:

```python
age = 15
can_vote = age >= 18
print(can_vote)
print(type(can_vote))
```

## 3. `=` vs. `==` — The Mistake Everyone Makes Once (30–40 min)

This is one of the most common beginner mistakes in *any* programming language, so let's tackle it head-on now.

- `=` **assigns** a value to a variable ("store this")
- `==` **compares** two values ("are these the same?")

Mixing them up causes very different (and sometimes confusing) results.

```python
# Assignment - stores 5 into x
x = 5
print(x)
```

```python
# Comparison - checks if x equals 5, gives back True/False
x = 5
print(x == 5)
print(x == 10)
```

Using `=` when you meant `==` (or vice versa) is a very common source of bugs. In fact, Python won't even let you use `=` inside certain places where a comparison is expected (like an `if` statement, starting next lesson) — it'll raise a `SyntaxError` to protect you.

## 4. Comparing Different Types & Strings (40–48 min)

You can compare strings too — Python checks them alphabetically (technically by character codes), and comparisons are **case-sensitive**:

```python
print("apple" == "apple")
print("apple" == "Apple")     # False! capital A vs lowercase a
print("apple" < "banana")     # True - "apple" comes before "banana" alphabetically
```

⚠️ Comparing very different types (like a string to a number) with `==` won't crash — it just always gives `False`, since they can never truly be "equal":

```python
print("5" == 5)     # False - a string is never equal to an int, even if it "looks" the same
print(5 == 5.0)     # True - int and float CAN be equal if the value matches
```

This is exactly why, back in Lesson 5, we had to convert `input()` results to `int`/`float` — comparing the raw string `"18"` to the number `18` would always be `False`, even though they "look" equal:

```python
age = input("Enter your age: ")   # this is a string!
print(age == 18)                   # always False, no matter what you type, unless converted

age = int(age)
print(age == 18)                   # works correctly now
```
