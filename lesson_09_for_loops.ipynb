## 1. Why Loops? (5–18 min)

Imagine you wanted to print "Hello!" five times. Right now, you'd have to write this:

````python
print("Hello!")
print("Hello!")
print("Hello!")
print("Hello!")
print("Hello!")
````

That works, but it's repetitive — and what if you wanted it 500 times instead of 5? Or what if the number of repetitions wasn't known until the program ran?

This is exactly the kind of problem loops solve. A **loop** repeats a block of code automatically, as many times as you tell it to.

````python
for i in range(5):
    print("Hello!")
````

Same result, one loop. Let's break down exactly what's happening.

---
## 2. `for` Loops with `range()` (18–30 min)

The basic shape of a `for` loop looks like this:

```python
for i in range(5):
    print(i)
```

- `for` starts the loop
- `i` is a **loop variable** — it changes value on every pass through the loop (you can name it anything, but `i` is a common convention)
- `range(5)` generates the numbers `0, 1, 2, 3, 4` — five numbers, starting at 0
- everything indented underneath runs once **per number** in that sequence

````python
for i in range(5):
    print(i)
````

Notice that `range(5)` counts from `0` to `4` — five numbers total, not `1` to `5`. This trips up almost everyone at first, so it's worth remembering: **`range(n)` gives you `n` numbers, starting at 0.**

You can use the loop variable inside the loop, not just print it:

````python
for i in range(5):
    print(f"This is loop number {i}")
````

````python
total = 0

for i in range(1, 6):
    total = total + i

print(f"The sum of 1 through 5 is {total}")
````

## 3. Looping Over Strings (30–42 min)

`range()` isn't the only thing you can loop over. You can loop directly over a string — Python will hand you one character at a time.

````python
name = "Python"

for letter in name:
    print(letter)
````

No `range()`, no indexes — Python just walks through the string character by character. This is often cleaner than looping with `range(len(name))` and indexing manually, though both work:

````python
name = "Python"

# Same result, using range() and indexing:
for i in range(len(name)):
    print(name[i])
````

A common use is counting or checking characters:

````python
word = "programming"
vowel_count = 0

for letter in word:
    if letter in "aeiou":
        vowel_count = vowel_count + 1

print(f"'{word}' has {vowel_count} vowels")
````

## 4. Variations on `range()` (42–48 min)

`range()` can take up to three arguments: `range(start, stop, step)`.

````python
# range(stop) -> starts at 0
for i in range(5):
    print(i)
````

````python
# range(start, stop) -> starts wherever you say
for i in range(2, 7):
    print(i)
````

````python
# range(start, stop, step) -> counts by the step
for i in range(0, 10, 2):
    print(i)
````

````python
# A negative step counts down
for i in range(10, 0, -1):
    print(i)
print("Liftoff! 🚀")
````

Just like slicing (from Lesson 3), `stop` in `range()` is never included. `range(2, 7)` gives you `2, 3, 4, 5, 6` — not `7`.
