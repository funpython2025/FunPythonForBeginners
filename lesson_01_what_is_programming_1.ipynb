## 1. What *is* Programming? (5–15 min)

A computer is incredibly powerful, but also incredibly **literal**. It only does exactly what you tell it — nothing more, nothing less, and never what you *meant* if that's not what you *said*.

**Programming** is the skill of writing precise, step-by-step instructions in a language the computer can understand. Those instructions are called **code**, and a set of steps that solves a problem is called an **algorithm**.

### 🥪 The Sandwich Algorithm
Imagine trying to explain "make a sandwich" to someone who has never seen one and takes everything 100% literally:

> 1. Pick up two slices of bread
> 2. Open the peanut butter jar
> 3. Put peanut butter on one slice using a knife
> 4. Put the two slices together

If you skip a step, or the order is wrong (spreading peanut butter *before* opening the jar!), it breaks. Code is exactly like this — order and precision matter enormously.

### Why Python?
**Python** is one of the most widely used programming languages in the world — it powers websites (Instagram, Spotify), video games, scientific research, and modern AI. It's a great first language because its instructions read almost like plain English.

## 2. Your First Commands: `print()` (15–30 min)

The very first thing every programmer learns is how to make the computer **output** something — display text on the screen. In Python, we do this with the `print()` function.

A **function** is a reusable, named command that performs an action. `print()` is a function that displays whatever you put inside its parentheses `()`.

```python
print("I am learning Python!")
```

Notice the quotation marks `" "` around the text. Text wrapped in quotes is called a **string** — it tells Python "treat this as literal text, not as code."

You can print numbers too, without quotes:

```python
print(42)
print(3.14)
```

You can print multiple things in one call by separating them with commas — Python automatically adds a space between them:

```python
print("My favorite number is", 7)
```

You can also call `print()` multiple times — each call starts on a new line:

```python
print("Line one")
print("Line two")
print("Line three")
```

### Single vs. double quotes
Python doesn't care whether you use `'single quotes'` or `"double quotes"` for strings — just be consistent within the same string:

```python
print('This works fine.')
print("So does this.")
```

## 4. Reading Errors — Your Friend, Not Your Enemy (40–50 min)

Every programmer — beginner or professional with 20 years of experience — sees errors constantly. An error just means Python is telling you *exactly* what's wrong. Learning to read error messages is one of the most valuable programming skills you can build.

Let's cause one on purpose. Run the cell below:

```python
print("this string is missing something"
```

This is a **`SyntaxError`** — it means the *grammar* of your code is broken (in this case, a missing closing parenthesis `)`). Python shows you:
- The **line** where it noticed the problem
- A `^` pointing roughly at where things went wrong
- The **type of error** and a short description

👉 **Fix it:** Add the missing `)` to the cell above and run it again.

Let's look at a different, very common error:

```python
print(hello)
```

This is a **`NameError`** — Python is telling you it has no idea what `hello` refers to. That's because `hello` isn't in quotes, so Python thinks it's supposed to be an existing variable — but we never created one with that name (we'll learn about variables in Lesson 2!).

👉 **Fix it:** Add quotes around `hello` in the cell above so it becomes a string, then run it again.

### Common beginner errors to watch for
| Error message | Usual cause |
|---|---|
| `SyntaxError` | Missing parenthesis, quote, or colon |
| `NameError` | Using a word Python doesn't recognize (often a missing quote) |
| `IndentationError` | Inconsistent spacing (we'll hit this a lot starting Lesson 7) |
