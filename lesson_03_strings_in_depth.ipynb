## 1. Concatenation vs. f-strings (5–15 min)

You can glue strings together with `+` — this is called **concatenation**. It only works between strings, and it does *not* add spaces automatically.

```python
first = "Py"
second = "thon"
print(first + second)

greeting = "Hello" + "World"   # no space!
print(greeting)

greeting_fixed = "Hello" + " " + "World"
print(greeting_fixed)
```

You can also "multiply" a string by a number to repeat it:

```python
print("ha" * 3)
print("-" * 20)
```

f-strings (from Lesson 2) are usually the cleaner choice when mixing text and variables, especially with numbers, since you don't need `str()`:

```python
name = "Sam"
wins = 4
print("Congrats " + name + ", you have " + str(wins) + " wins!")   # concatenation - clunky
print(f"Congrats {name}, you have {wins} wins!")                    # f-string - clean
```

## 2. Indexing & Slicing (15–30 min)

A string is really a sequence of characters, and each character has a position called an **index**. Python starts counting at **0**, not 1!

```python
word = "PYTHON"
#        012345

print(word[0])   # P
print(word[1])   # Y
print(word[5])   # N
```

Negative indexes count from the **end** of the string, starting at `-1`:

```python
word = "PYTHON"
print(word[-1])   # N (last character)
print(word[-2])   # O (second to last)
```

### Slicing
You can grab a *range* of characters using `word[start:stop]`. This gives you everything from `start` up to (but **not including**) `stop`.

```python
word = "PYTHON"

print(word[0:3])   # P Y T -> "PYT"
print(word[2:5])   # T H O -> "THO"
print(word[:3])    # leaving out start means "from the beginning"
print(word[3:])    # leaving out stop means "to the end"
print(word[:])     # the whole string
```

You can also find how many characters are in a string with `len()`:

```python
word = "PYTHON"
print(len(word))
```

⚠️ Strings are **immutable** — once created, you can't change a character in place. `word[0] = "B"` would cause an error. Instead, you build a *new* string, often using slicing:

```python
word = "PYTHON"
new_word = "B" + word[1:]
print(new_word)
```

## 3. String Methods (30–45 min)

Strings come with many built-in **methods** — mini functions attached to the string itself, called with a dot: `string.method()`. Here are ones you'll use constantly:

```python
text = "  Hello, World!  "

print(text.upper())        # HELLO, WORLD!
print(text.lower())        # hello, world!
print(text.strip())        # removes leading/trailing whitespace
print(text.replace("World", "Python"))
```

```python
sentence = "python is fun and python is powerful"

print(sentence.count("python"))     # how many times "python" appears
print(sentence.find("fun"))         # index where "fun" starts
print(sentence.split())             # splits into a list of words (more on lists in Lesson 11!)
```

```python
# Checking things about a string
password = "Secure123"

print(password.isdigit())     # False - contains letters
print(password.isalpha())     # False - contains numbers
print(password.startswith("Sec"))
print(password.endswith("123"))
```

💡 A key idea: **methods don't change the original string** — strings are immutable, remember? They return a *new* string. If you want to keep the change, store it in a variable.

```python
name = "alex"
print(name.upper())   # prints ALEX
print(name)            # name is still "alex" - unchanged!

name = name.upper()   # now we save it back
print(name)            # ALEX
```
