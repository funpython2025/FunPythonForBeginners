## 1. The `if` Statement & Indentation (5–18 min)

An `if` statement runs a block of code **only if** a condition is `True`. Here's the basic shape:

```python
if condition:
    # this runs only if condition is True
```

````python
age = 15

if age >= 13:
    print("You're a teenager!")
````

Let's break down the syntax carefully, because every piece matters:
- `if` starts the statement
- `age >= 13` is the **condition** — it must evaluate to `True` or `False`
- The line ends with a **colon** `:`
- The next line is **indented** (4 spaces) — this is called the **body** of the `if` statement

### ⚠️ Indentation is not optional in Python
Unlike many other languages, Python uses indentation to know which lines belong inside the `if` statement. Forgetting to indent causes an `IndentationError`. Try running the broken cell below:

````python
age = 15

if age >= 13:
print("You're a teenager!")
````

👉 **Fix it:** Add 4 spaces (or one Tab, Jupyter usually handles this for you automatically after a colon) before `print(...)` in the cell above.

What happens if the condition is `False`? The indented block is simply **skipped** — nothing happens, the program just moves on:

````python
age = 10

if age >= 13:
    print("You're a teenager!")

print("Program continues here either way.")
````

## 2. `else` — the "Otherwise" Branch (18–28 min)

What if we want to do something specific when the condition is `False`, instead of just skipping it? That's what `else` is for.

````python
age = 10

if age >= 13:
    print("You're a teenager!")
else:
    print("You're not a teenager yet.")
````

Exactly one of the two blocks will run — never both, never neither. `else` also needs its own colon, and its body is indented the same way as the `if` body.

## 3. `elif` — Checking Multiple Conditions (28–40 min)

Real decisions often have more than two outcomes. `elif` (short for "else if") lets you check additional conditions in order, and you can chain as many as you need.

````python
grade = 85

if grade >= 90:
    print("A")
elif grade >= 80:
    print("B")
elif grade >= 70:
    print("C")
else:
    print("Needs improvement")
````

⚠️ **Order matters, and Python stops at the first `True` condition.** Even though `85 >= 70` is also true, Python never even checks that line — it already found a match at `grade >= 80` and skipped the rest. This is why we check from highest to lowest here.

````python
# What would happen if we ordered these the OTHER way? Try it and see why order matters!
grade = 85

if grade >= 70:
    print("C")
elif grade >= 80:
    print("B")
elif grade >= 90:
    print("A")
else:
    print("Needs improvement")
````

You can have as many `elif` blocks as you want, but only **one** `if` (to start) and **at most one** `else` (at the end, optional).

````python
day = "Wednesday"

if day == "Saturday":
    print("It's the weekend!")
elif day == "Sunday":
    print("It's the weekend!")
elif day == "Wednesday":
    print("Halfway through the week!")
else:
    print("Just a regular weekday.")
````

## 4. Nested `if` Statements (40–48 min)

You can put an `if` statement **inside** another one — this is called **nesting**, and it's how you handle "and this other thing must also be true, but only in this specific case" logic.

````python
age = 16
has_permission = True

if age >= 13:
    print("You're a teenager.")
    if has_permission:
        print("You have permission to attend.")
    else:
        print("You need a permission slip.")
else:
    print("You're too young for this event.")
````

Notice the **double indentation** — the inner `if`/`else` is indented once inside the outer `if`'s body.

💡 Often, nested `if` statements can be simplified using `and` (from Lesson 6/8 preview) — we'll practice that combination next lesson. For now, nesting is a useful tool to have.
