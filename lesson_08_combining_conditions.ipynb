## 1. `and` — Both Must Be True (5–18 min)

`and` combines two conditions into one. The combined result is `True` **only if both sides are `True`**.

```python
age = 16
has_ticket = True

print(age >= 13 and has_ticket)   # True and True -> True
```

```python
age = 10
has_ticket = True

print(age >= 13 and has_ticket)   # False and True -> False
```

This is exactly the nested `if` from last lesson, simplified into one line:

```python
age = 16
has_permission = True

# Last lesson's nested version:
if age >= 13:
    if has_permission:
        print("You can attend!")

print("---")

# Today's cleaner version, using and:
if age >= 13 and has_permission:
    print("You can attend!")
```

You can chain more than two conditions with `and`:

```python
age = 16
has_ticket = True
arrived_on_time = True

if age >= 13 and has_ticket and arrived_on_time:
    print("Welcome in!")
else:
    print("Sorry, you can't enter.")
```

## 2. `or` — At Least One Must Be True (18–28 min)

`or` also combines two conditions, but the result is `True` if **at least one** side is `True` — only fails if *both* are `False`.

```python
day = "Saturday"

if day == "Saturday" or day == "Sunday":
    print("It's the weekend!")
else:
    print("It's a weekday.")
```

```python
# Try changing "day" above to "Monday" and re-running to see the other branch
```

A common beginner mistake: writing `day == "Saturday" or "Sunday"`. This does **not** do what you'd think — you must repeat the full comparison on both sides!

```python
day = "Monday"

# WRONG - this is checking "day == Saturday" OR "Sunday" (which is always truthy!)
print(day == "Saturday" or "Sunday")

# RIGHT - repeat the full comparison each time
print(day == "Saturday" or day == "Sunday")
```

## 3. `not` — Flipping a Boolean (28–36 min)

`not` simply flips `True` to `False`, and `False` to `True`.

```python
is_raining = True
print(not is_raining)    # False

is_sunny = False
print(not is_sunny)      # True
```

`not` is especially handy to make conditions read more naturally in English:

```python
has_permission = False

if not has_permission:
    print("You need permission to proceed.")
```

```python
username = "guest"

if username != "admin":
    print("Access restricted.")

# Same idea, using not instead of !=
if not (username == "admin"):
    print("Access restricted.")
```

## 4. Combining Operators & Truth Tables (36–46 min)

You can mix `and`, `or`, and `not` together, and use parentheses to group them (just like math order of operations from Lesson 4!).

```python
age = 20
has_id = True
is_banned = False

can_enter = (age >= 18 and has_id) and not is_banned
print(can_enter)
```

### Truth tables
A **truth table** shows every possible combination of `True`/`False` inputs and the resulting output. Here's `and` and `or` side by side:

| A | B | A and B | A or B |
|---|---|---------|--------|
| True | True | True | True |
| True | False | False | True |
| False | True | False | True |
| False | False | False | False |

Notice: `and` is only `True` in one row (both True). `or` is `False` in only one row (both False).

```python
# Let's verify the truth table ourselves with a loop... well, we don't know loops yet!
# For now, let's just check each combination manually:
print(True and True)
print(True and False)
print(False and True)
print(False and False)
print("---")
print(True or True)
print(True or False)
print(False or True)
print(False or False)
```
