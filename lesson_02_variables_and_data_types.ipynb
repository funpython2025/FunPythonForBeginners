## 1. Variables — Giving Values a Name (5–20 min)

So far, every time we wanted to reuse "Alex" or `42`, we had to retype it. A **variable** is like a labeled box: you store a value inside it and can reuse that value later just by using its name.

```python
name = "Alex"
age = 15

print(name)
print(age)
```

The `=` sign here doesn't mean "equals" like in math class — it means **"store this value under this name."** This is called **assignment**. Python evaluates the right side first, then stores the result in the name on the left.

Variables can be updated at any time — the box's label stays the same, but you can swap what's inside:

```python
score = 0
print("Starting score:", score)

score = score + 10
print("Updated score:", score)

score = score + 5
print("Updated score again:", score)
```

Notice `score = score + 10` — this reads the *current* value of `score`, adds 10 to it, and stores the result back into `score`. This pattern (read, change, store back) is extremely common.

```python
# There's also a shortcut for that pattern:
score = 0
score += 10   # same as: score = score + 10
print(score)

score -= 3    # same as: score = score - 3
print(score)
```

## 2. Data Types (20–35 min)

Every value in Python has a **type** — it tells Python (and you) what *kind* of value it is, and what you're allowed to do with it. The four basic types you'll use constantly:

| Type | Meaning | Example |
|------|---------|---------|
| `str` (string) | Text | `"hello"` |
| `int` (integer) | Whole number | `42` |
| `float` | Decimal number | `3.14` |
| `bool` (boolean) | True or False | `True` |

You can check any value's type with the `type()` function:

```python
print(type("hello"))
print(type(42))
print(type(3.14))
print(type(True))
```

### Types matter!
Python treats types very differently — mixing them incorrectly causes errors. Watch what happens when we try to combine text and a number with `+`:

```python
age = 15
print("I am " + age + " years old")
```

That's a `TypeError` — Python won't automatically turn a number into text for you. You have two options: convert the number to a string with `str()`, or (better, coming up next) use an f-string.

```python
age = 15
print("I am " + str(age) + " years old")
```

### Converting between types
You can convert values between types using `str()`, `int()`, and `float()`:

```python
# text to whole number
num_text = "10"
num = int(num_text)
print(num + 5)

# whole number to decimal
whole = 7
decimal = float(whole)
print(decimal)

# decimal to whole number (this cuts off, doesn't round!)
pi_ish = 3.9
print(int(pi_ish))
```

⚠️ Note: `int(3.9)` gives `3`, not `4` — converting a float to an int **truncates** (chops off the decimal) rather than rounding. Use `round()` if you want proper rounding: `round(3.9)` gives `4`.

## 3. Naming Rules & f-strings (35–45 min)

### Variable naming rules
- Must start with a letter or underscore (not a number)
- Can contain letters, numbers, and underscores — no spaces or symbols
- Case-sensitive: `score` and `Score` are different variables
- Can't use Python's reserved words (like `print`, `if`, `for`)

By convention, Python variable names use `snake_case` — lowercase words separated by underscores, like `favorite_color` or `total_score`.

```python
# Valid names
favorite_color = "blue"
total_score = 100
_hidden = "starts with underscore, valid but unusual"

# This would be invalid if uncommented: 2fast = "nope"  -- can't start with a number
print(favorite_color, total_score)
```

### f-strings: the clean way to combine text and variables
Instead of `print("I am " + str(age) + " years old")`, put an `f` right before the opening quote, and place variables directly inside curly braces `{}`. Python fills them in automatically — no `str()` needed, and it works for any type.

```python
name = "Alex"
age = 15
gpa = 3.8

print(f"Hi, I'm {name}. I'm {age} years old with a GPA of {gpa}.")
```

You can even put simple math inside the curly braces:

```python
age = 15
print(f"Next year I'll be {age + 1}.")
```
